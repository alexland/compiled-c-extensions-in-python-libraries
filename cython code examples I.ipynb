```python
%%cython -af

import random as RND
cimport cython
from cython cimport cdivision
from cython cimport boundscheck, wraparound, cdivision

# import directly from c libraries
from libc.math cimport log10, pow, floor
from libc.stdlib cimport malloc

import numpy as NP
cimport numpy as NP


@cdivision(True)
cdef inline int get_digit(int num, int radix, int pos):
    '''
    returns:
    pass in
    '''
    cdef:
        double a
    a = floor(num / pow(radix, pos))
    res = a % radix
    return <int>res


@cdivision(True)
@cython.boundscheck(False)
@cython.wraparound(False)
@cython.profile(False)
cpdef inline prefix_sum(int[:] a):
    '''
    returns:
    pass in:
    '''
    cdef:
        unsigned short i
        unsigned int len_a = a.size
    for i in range(1, len(a)):
        a[i] += a[i-1]
    return a



@cdivision(True)
@cython.boundscheck(False)
@cython.wraparound(False)
@cython.profile(False)
cpdef radix_sort(long[:] array_in, unsigned short radix):
    '''
    returns:
    pass in:
    '''
    cdef:
        unsigned int i
        int e, f
        unsigned short num_scans
        int max_val = -32000
    
    len_array_in = len(array_in)

    for i in range(len_array_in):
        if array_in[i] > max_val:
            max_val = array_in[i]
            
    num_scans = <unsigned short>floor(log10(max_val)+1)
    array_out = NP.zeros_like(array_in)
    
    for pos in range(num_scans):
        count = [0] * radix
        for e in array_in:
            digit = get_digit(e, radix, pos)
            count[digit] += 1
        
        count = prefix_sum(count)
        
        for f in reversed(array_in):
            digit = get_digit(f, radix, pos)
            count[digit] -= 1
            new_pos = count[digit]
            array_out[new_pos] = f
        
        array_in = array_in.copy()
    return array_out
            
    
    


def wrap_cdef(array_in, radix):
    return radix_sort(array_in, radix)
```
