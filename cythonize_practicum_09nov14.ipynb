```python
%%cython -af

cimport cython
from cython cimport boundscheck, wraparound, cdivision
from libc.math cimport exp, sqrt, pow, log, erf
from libc.stdlib cimport malloc
cimport numpy as NP
import numpy as NP


@boundscheck(False)
@wraparound(False)
@cdivision(True)
cpdef fnx():
    '''
    views a C array w/ a typed memoryview
    by assigning the arrray to the memoryview;
    
    'A' below is static, so stack-allocated & so
    tied to the fn from which it was allocated
    '''
    # declare & init C array, A
    cdef int A[1000][1000]
    
    # declare mv1 as a C-contiguous memoryview of a
    cdef int[:,::1] mv_a = A
    
    # initialize the array to all 1s
    mv_a[...] = 1


cpdef fnx1(size_t, N, size_t M):
    '''
    same as above except C array is dyamically allocated
    so cython doesn't know its extent;
    
    'A' below is heap-allocated so not tied to the fn 
    from which it was allocated
    '''
    # declare a C array as a long pointer, bound to 'A'
    cdef long *A = <long*>malloc(N * M * sizeof(long))
    
    # assign A to a typed memoryview, as in you do for
    # fixed-size arrays, using typed memoryview cast
    cpdef long[:,::1]mv_a = <long[:N, :M]>A
    
    

    
@boundscheck(False)
@wraparound(False)
@cdivision(True)
def mv_sum(int[:,::1] mv):
    cdef unsigned int N, M, i, j
    cdef long s = 0
    N = mv.shape[0]
    M = mv.shape[1]
    for i in range(N):
        for j in range(M):
            s += mv[i, j]
    return s
```

### insertion sort

```python
def insertion_sort(alist):
    alist_len = len(alist)
    # traverse the list to be sorted
    for i in range( 1, alist_len):
        tmp = alist[i]
        k = i
        # if an item is smaller than the one before it
        while k > 0 and tmp < alist[k - 1]:
            # replace 2nd item w/ 1st (larger)
            alist[k] = alist[k - 1]
            k -= 1
        # complete the swap: replace 1st item w/ 2nd (smaller)
        alist[k] = tmp
```

### replacing singe-method class w/ function using a closure

```python
from urllib.request import urlopen

class UrlTemplate:
    
    def __init__(self, template):
        self.template = template
        
    def open(self, **kwargs):
        return urlopen(self.template.format_map(kwargs))
        
```

```python
yahoo_finance_data = UrlTemplate('http://finance.yahoo.com/d/quotes.csv?s={names}&f={fields}')

for line in yahoo_finance_data.open(names='FB', fields='sl1c1v'):
    D = line.decode('utf-8')
```
